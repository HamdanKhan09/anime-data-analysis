```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Matploitlib Basics**

```python
x=np.linspace(0,5,11)
x
y= x**2
y
```

```python
plt.plot(x,y)
plt.title('Hello')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y**2)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x**2,y)
```

```python
fig = plt.figure(dpi = 100)
axis1 = fig.add_axes([.1,.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.3,0.3,0.3])
axis2.plot(x,y)
```

```python
plt.scatter(x,y)
```

```python
plt.boxplot(x)
```

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])
ax.plot(x,y)
plt.savefig("basicplot.png")
```
